```python
# STORY 1: UNIVARIATE ANALYSIS
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load data
df = pd.read_csv('../data/insurance_claims.csv')
print("Dataset:", df.shape)
```

```python
# BAR CHART FOR FRAUD DETECTION (was histogram)
plt.figure(figsize=(10, 6))
fraud_counts = df['fraud_reported'].value_counts().sort_index(ascending=False)
plt.bar(['Y', 'N'], fraud_counts.values, color=['skyblue', 'orange'], edgecolor='black')
plt.title('Bar Chart: Fraud Detection Distribution')
plt.ylabel('Frequency')
plt.savefig('../outputs/story1/fraud_detection.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# HISTOGRAM FOR AGE DISTRIBUTION

plt.figure(figsize=(8, 6))
plt.hist(df['age'], bins=10, edgecolor='black', color='lightcoral', alpha=0.7)
plt.title('Histogram: Age Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.3)
plt.savefig('../outputs/story1/age_distribution.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# PIE CHART FOR DAMAGE SEVERITY (4 categories)
plt.figure(figsize=(10, 8))

damage_counts = df['incident_severity'].value_counts()

colors = {
    'Minor Damage': '#FF6B6B',      
    'Major Damage': '#4ECDC4',        
    'Trivial Damage': '#45B7D1',    
    'Total Loss': '#F7DC6F'         
}

labels = ['Minor Damage', 'Major Damage', 'Trivial Damage', 'Total Loss']
order = ['Minor Damage', 'Major Damage', 'Trivial Damage', 'Total Loss']
colors_list = [colors.get(label, '#95A5A6') for label in order]

damage_filtered = damage_counts.reindex(order, fill_value=0)
plt.pie(damage_filtered.values, 
        labels=labels, 
        colors=colors_list,
        autopct='%1.1f%%', 
        startangle=90,
        explode=(0.05, 0, 0, 0),  
        shadow=True)
plt.title('Damage Severity Distribution', fontsize=14, fontweight='bold')
plt.savefig('../outputs/story1/pie_damage_visualization.png', dpi=300, bbox_inches='tight')
plt.show()
```
